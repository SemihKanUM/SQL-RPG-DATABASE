```python
## Type of chat =  -- 'One-on-One' or 'Group'
```

```python
%%sql 
CREATE TABLE Chat (
    ID INT AUTO_INCREMENT,
    Type VARCHAR(255),
    PRIMARY KEY (ID)
);
```

```python
%%sql
CREATE TABLE Message (
    ID INT AUTO_INCREMENT,
    Content TEXT,
    SenderID INT,
    Timestamp DATETIME,
    ChatID INT,
    PRIMARY KEY (ID),
    FOREIGN KEY (SenderID) REFERENCES `Character`(ID),
    FOREIGN KEY (ChatID) REFERENCES Chat(ID)
);
```

```python
%%sql
ALTER TABLE `Character`
ADD COLUMN ChatID INT,
ADD FOREIGN KEY (ChatID) REFERENCES Chat(ID);
```

```python
## Player to Player private chat outside of the game
```

```python
%%sql
ALTER TABLE Player
ADD COLUMN PrivateChatID INT,
ADD FOREIGN KEY (PrivateChatID) REFERENCES Chat(ID);
```

```python
%%sql
CREATE TABLE ChatLog (
    ID INT AUTO_INCREMENT,
    ChatID INT,
    SenderID INT,
    Content TEXT,
    Timestamp DATETIME,
    PRIMARY KEY (ID),
    FOREIGN KEY (ChatID) REFERENCES Chat(ID),
    FOREIGN KEY (SenderID) REFERENCES `Character`(ID)
);
```

```python
import mysql.connector

# Replace these with your actual database credentials
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': '12345678',
    'database': 'Aetheria',
}

# SQL statement to create the trigger
create_trigger_sql = """
CREATE TRIGGER after_insert_chat_message
AFTER INSERT
ON Message FOR EACH ROW
BEGIN
    INSERT INTO ChatLog (ChatID, SenderID, Content, Timestamp)
    VALUES (NEW.ChatID, NEW.SenderID, NEW.Content, NEW.Timestamp);
END
"""

# Connect to the database and execute the SQL statement
try:
    connection = mysql.connector.connect(**db_config)
    cursor = connection.cursor()
    cursor.execute(create_trigger_sql)
    connection.commit()
    print("Trigger created successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
```
